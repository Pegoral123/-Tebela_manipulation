```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv"
dados = pd.read_csv(url, sep=",")
dados
```

```python
df = dados.fillna(0)
valores_remover =  df[df['Notas'] == 0].index
dados.drop(valores_remover, axis = 0, inplace= True)
```

```python
index_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index
dados.drop(index_remover, axis = 0, inplace=True)
```

```python
df_aprovados = dados.query('Aprovado == True')
```

```python
df_aprovados.to_csv('alunos_aprovados.csv', index = False, sep = ';')
```
