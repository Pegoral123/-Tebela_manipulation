# ***INTRODUÇÃO:***

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv"
dados = pd.read_csv(url, sep=",")
dados
```

```python
df = dados.fillna(0) #estamos tratando dados nulos e subestituindo pelo valor 0 com metoodo fillna
valores_remover =  df[df['Notas'] == 0].index #pegando o index para exluir a linhas com notas 0
dados.drop(valores_remover, axis = 0, inplace= True) #excluindo. AXIS = O exclui a linha e AXIS =1 exclui a coluna.
```

```python
index_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index #aqui estamos localizando o index dos alunos informados e armazendo para excluir
dados.drop(index_remover, axis = 0, inplace=True) #usando o metodo drop para excluir o alunos alice e carlos do data frame
```

```python
df_aprovados = dados.query('Aprovado == True') #aqui estamos pegando e armazenado os alunos aprovados
```

```python
df_aprovados.replace(7.0, 8.0) #o metodo replace ele usa o primeiro parametro como index para atualizar pelo segundo
```

```python
df_aprovados.to_csv('alunos_aprovados.csv', index = False, sep = ';')
```

## **Desafio: 4**

1) Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.

```python
df_alunos = dados #nomeando data frema para alunos
df_alunos.head()
```

```python
df_alunos['Pontos_extras'] = df_alunos['Notas'] * 0.40 #criando coluna pontos extras e acrecentando valor
df_alunos
```

2) Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

```python
df_alunos['Notas_finais'] = df_alunos['Notas'] + df_alunos['Pontos_extras']
df_alunos.head()
```

3) Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
False: caso o aluno esteja reprovado (nota final deve ser menor que 6)

```python
df_alunos["Aprovados_final"] = df_alunos['Notas_finais'].apply(lambda x: True if x >= 6 else False )
df_alunos.head()
```

4) Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

```python
df_alunos_aprovados = df_alunos.query('Aprovado == False & Aprovados_final == True')
df_alunos_aprovados.head()
```
